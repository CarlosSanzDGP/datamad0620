```python
import pandas as pd
```

```python


df = pd.read_sql_query("""
SELECT au.au_id  author_id,
	au_lname Last_name,
    au_fname First_name,
    t.title Title,
    p.pub_name Publisher
FROM authors au
	INNER JOIN titleauthor tau
    ON au.au_id = tau.au_id
		LEFT JOIN titles t
		ON tau.title_id = t.title_id
	LEFT JOIN publishers p
	ON t.pub_id = p.pub_id
    ORDER BY Last_name
""", engine)

```
